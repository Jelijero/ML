```python
import gradio as gr
import speech_recognition as sr
from pydub import AudioSegment
import PyPDF2
```

```python
from gtts import gTTS
from IPython.display import Audio
```

```python
from transformers import AutoTokenizer,  AutoModelForSeq2SeqLM
```

```python
# Open the PDF file
with open("/content/drive/MyDrive/Docs.pdf", "rb") as file:
    reader = PyPDF2.PdfReader(file)
    document = ""

    # Extract text from each page
    for page in reader.pages:
        document += page.extract_text()
```

```python
document
```

```python
tokenizer = AutoTokenizer.from_pretrained("valhalla/t5-small-qa-qg-hl")
model =  AutoModelForSeq2SeqLM.from_pretrained("valhalla/t5-small-qa-qg-hl")
```

```python


def speech_to_text(audio_file):
    # Convert the uploaded audio file to WAV format (Gradio automatically handles the audio as a WAV file)
    recognizer = sr.Recognizer()

    # Load the audio file
    with sr.AudioFile(audio_file) as source:
        audio_data = recognizer.record(source)

    # Recognize speech using Google Web Speech API
    try:
        text = recognizer.recognize_google(audio_data)

    except sr.UnknownValueError:
        return "Google Speech Recognition could not understand the audio"
    except sr.RequestError as e:
        return f"Could not request results from Google Speech Recognition service; {e}"
    # Prepare the input text in T5 format
    input_text = f"question: {text}? context:{document} "

    # Tokenize the input
    input_ids = tokenizer(input_text, return_tensors="pt").input_ids
    outputs = model.generate(
    input_ids,
    max_length=130,
    min_length=30,
    top_k=10,
    length_penalty=1,
    num_beams=8,
    top_p=1,
    temperature=1,
    early_stopping=True,
    repetition_penalty=10.0
    )

    answer = tokenizer.decode(outputs[0], skip_special_tokens=True)
    tts = gTTS(answer)
    tts.save("hello.mp3")

    return "hello.mp3"


# Create a Gradio interface

```

```python
interface = gr.Interface(
    fn=speech_to_text,
    inputs=gr.Audio(type="filepath"),  # Expect an uploaded audio file
    outputs=gr.Audio(type="filepath"),
    title="Speech to Speech Chatbot"

)

# Launch the interface
interface.launch(debug=True)
```
